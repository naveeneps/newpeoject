‚û°Ô∏è  A Tuple is an Immutable Data Structure , in simple words it means once a tuple is created no modifications/manipulations are allowed in tuple elements .

in other words

```python
## Packing and Unpacking
# --
```

```python
# this is packing

ice_creme =  ('chocolate' ,'rum-raisin' ) # ‚û°Ô∏è  A Tuple is Represented using ( )
```

```python
# extracting tuple elements

# traditional method

f1 = ice_creme[0]
f2 = ice_creme[1]

print(f1,f2)
```

```python
# smart: unpacking
s1 ,s2 = ice_creme
print( s1, s2 )
```
