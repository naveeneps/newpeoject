# Characteristics of Dictionary
---
<pre>
A Single Element in Dictionary contains Key/Value

Dictionary is Mutable and Ordered Data Structure.

You cannot Repeat a Key in Dictionary . ( can repeat value)

you can Only Use Immutable DataTypes as KEY , Mutable DataTypes as Key aren't Allowed.

eg.

Immutable Types : int /str/ float /bool / tuple

mutable types : set/ dictionary / list

eg.2

  { 'a' : 'apple' }  # ‚úÖ 'a' is String is immutable hence valid.

  { ['a'] : 'apple' } # ‚ùå ['a'] is List is Mutable hence Invalid.

```python
# you are asked to Accept  5 Distance values in Miles , and Convert Them into Kilometers , use Miles  as Key, use Kilometer as Value.
```

```python
distance = { }

for _ in range( 1,6) :

  miles = int (input('enter distance in Miles '))

  # syntax:       dict [ keyhere ] =  value_Here

  distance [ miles ] = miles * 1.6


print('--------Report----------')

print(distance)
```

```python
# You are provided with Student Scores

scores = [ 50 , 32 , 51 , 24 , 29 , 61 ]

# you are asked to generate a Dictionary where

# key : score

# value : 'Pass' or 'Fail'     ,            any score < 40 is 'Fail' otherwise 'Pass'
```

```python
scores = [ 50 , 32 , 51 , 24 , 29 , 61 ]
```

```python
reportCard = {}


for score  in scores :

  if score < 40 :

    reportCard[ score ] = 'Fail'

  else:

    reportCard[ score ] = 'Pass'


print( reportCard )
```
